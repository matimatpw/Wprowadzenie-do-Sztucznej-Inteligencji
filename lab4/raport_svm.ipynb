## Badanie wpływu hiperparametrów na algorytm SVM

### Hiperparemetry:
- C (Parametr reguralyzacji)
- Kernel Function (Metoda jądrowa -> Liniowa lub RBF)
- Sigma (Przy metodzie jądrowej RBF)
- Learning Rate (Rozmiar kroku używany w gradiencie)
- Iterations (Iteracje przez zbior danych podczas trenowania)

Wyniki Przedstawione będą w tabelkach.

```python
from enum import Enum
from svm import SVM, optim_params, PrepareData, kernel_types
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
import pandas as pd
```

```python
data = PrepareData()
X_train, X_test, y_train, y_test = train_test_split(np.array(data.X),np.array(data.y), test_size=0.2, random_state=18)
y_train, y_test = data.wrap_targets(y_train,y_test)
```

# Funkcja jądrowa - liniowa

### Zmienna C

```python
accuracy = []
c_params = [0.1, 2, 10]
for c in c_params:
    op_linear = optim_params(C=c, kernel=kernel_types.LINEAR,lr=1e-3, iters=100)
    svm = SVM(op_linear)
    svm.fit(X_train, y_train)

    predicts = svm.predict(X_test)
    accuracy.append(accuracy_score(y_test, predicts))


result = pd.DataFrame(
    {
        'C': c_params,
        'learning_rate': [1e-3 for _ in range(3)],
        'iterations': [100 for _ in range(3)],
        'accuracy': accuracy })

result
```

### Wspołczynnik uczenia 
(Learning rate)

```python
accuracy = []
lr_params = [1e-10, 1e-4, 1]
for lrate in lr_params:
    op_linear = optim_params(C=1.0, kernel=kernel_types.LINEAR,lr=lrate, iters=100)
    svm = SVM(op_linear)
    svm.fit(X_train, y_train)

    predicts = svm.predict(X_test)
    accuracy.append(accuracy_score(y_test, predicts))


result = pd.DataFrame(
    {
        'C': [1.0 for _ in range(3)],
        'learning_rate': lr_params,
        'iterations': [100 for _ in range(3)],
        'accuracy': accuracy })

result
```

### Iteracje
Iterations

```python
accuracy = []
iters_params = [100, 300, 500]
for iter in iters_params:
    op_linear = optim_params(C=1.0, kernel=kernel_types.LINEAR,lr=1e-3, iters=iter)
    svm = SVM(op_linear)
    svm.fit(X_train, y_train)

    predicts = svm.predict(X_test)
    accuracy.append(accuracy_score(y_test, predicts))


result = pd.DataFrame(
    {
        'C': [1.0 for _ in range(3)],
        'learning_rate': [1e-3 for _ in range(3)],
        'iterations': iters_params,
        'accuracy': accuracy })

result
```

# Funkcja jądrowa RBF

### Zmienna C

```python
accuracy = []
c_params = [0.1, 2, 10]
for c in c_params:
    op_rbf = optim_params(C=c, kernel=kernel_types.RBF,lr=1e-3, iters=100, sigma=0.5)
    svm = SVM(op_rbf)
    svm.fit(X_train, y_train)

    predicts = svm.predict(X_test)
    accuracy.append(accuracy_score(y_test, predicts))


result = pd.DataFrame(
    {
        'C': c_params,
        'sigma': [0.5 for _ in range(3)],
        'learning_rate': [1e-3 for _ in range(3)],
        'iterations': [100 for _ in range(3)],
        'accuracy': accuracy })

result
```

### Zmienna sigma
W funkcji rbf

```python
accuracy = []
sigma_params =  [0.1  , 0.5 , 2  ]
for sig in sigma_params:
    op_rbf = optim_params(C=1.0, kernel=kernel_types.RBF,lr=1e-4, iters=100, sigma=sig)
    svm = SVM(op_rbf)
    svm.fit(X_train, y_train)

    predicts = svm.predict(X_test)
    accuracy.append(accuracy_score(y_test, predicts))


result = pd.DataFrame(
    {
        'C': [1.0 for _ in range(3)],
        'sigma': sigma_params,
        'learning_rate': [1e-4 for _ in range(3)],
        'iterations': [100 for _ in range(3)],
        'accuracy': accuracy })


result
```

### Współczynik uczenia
Learning rate

```python
accuracy = []
lr_params = [1e-10, 1e-4, 1]
for lrate in lr_params:
    op_rbf = optim_params(C=1.0, kernel=kernel_types.RBF,lr=lrate, iters=100, sigma=0.5)
    svm = SVM(op_rbf)
    svm.fit(X_train, y_train)

    predicts = svm.predict(X_test)
    accuracy.append(accuracy_score(y_test, predicts))


result = pd.DataFrame(
    {
        'C': [1.0 for _ in range(3)],
        'sigma': [0.5 for _ in range(3)],
        'learning_rate': lr_params,
        'iterations': [100 for _ in range(3)],
        'accuracy': accuracy })

result
```

### Iteracje
Iterations

```python
accuracy = []
iters_params = [100  , 300 , 500]
for iter in iters_params:
    op_rbf = optim_params(C=1.0, kernel=kernel_types.RBF,lr=1e-3, iters=iter, sigma=0.5)
    svm = SVM(op_rbf)
    svm.fit(X_train, y_train)

    predicts = svm.predict(X_test)
    accuracy.append(accuracy_score(y_test, predicts))

result = pd.DataFrame(
    {
        'C': [1.0 for _ in range(3)],
        'sigma': [0.5 for _ in range(3)],
        'learning_rate': [1e-3 for _ in range(3)],
        'iterations': iters_params,
        'accuracy': accuracy})

result
```
